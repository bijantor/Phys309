```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

def ct_prime_axis(x_vals, v):

    beta = v / constants.c
    gamma = 1 / np.sqrt(1 - beta**2)
    return beta * x_vals / (1 - beta**2)


x_vals = np.linspace(-1, 1, 100)

"""tried to use 0.5c as v"""
v = 0.5 * constants.c
ct_vals = ct_prime_axis(x_vals, v)


plt.figure(figsize=(5,5))
plt.xlim([-1,1])
plt.ylim([-1,1])


plt.annotate(text='', xy=(-1,0), xytext=(1,0), arrowprops=dict(arrowstyle='<->'))
plt.annotate(text='', xy=(0,-1), xytext=(0,1), arrowprops=dict(arrowstyle='<->'))


plt.text(1.02, 0.02, r"$+x$", horizontalalignment='center', verticalalignment='bottom')
plt.text(-1.02, 0.02, r"$-x$", horizontalalignment='center', verticalalignment='bottom')
plt.text(-0.04, 1.02, r"$+ct$", horizontalalignment='right', verticalalignment='center')
plt.text(-0.04, -1.02, r"$-ct$", horizontalalignment='right', verticalalignment='center')


plt.plot(x_vals, ct_vals, 'r', linestyle='--', label=r"$ct'$ axis")


plt.legend()
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants


def ct_prime_axis(x_vals, v):

    beta = v / constants.c
    return beta * x_vals / (1 - beta**2)

def x_prime_axis(x_vals, v):

    beta = v / constants.c
    return beta * x_vals


x_vals = np.linspace(0, 2e10, num=201, endpoint=True) - 1e10


v = 0.2 * constants.c


ct_vals_prime = ct_prime_axis(x_vals, v)
ct_vals_x_prime = x_prime_axis(x_vals, v)


fig, ax = plt.subplots(figsize=(7,7))


ax.plot(x_vals, ct_vals_prime, 'r', linestyle='--', label=r"$ct'$ axis")
ax.plot(x_vals, ct_vals_x_prime, 'r', linestyle='--', label=r"$x'$ axis")


ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.set_aspect('equal')

ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])


ax.set_xlabel(r"$x$ (m)", fontsize=14, color="black")
ax.set_ylabel(r"$ct$ (m)", fontsize=14, color="black")


ax.text(8e9, 5e9, r"$ct'$", color='red', fontsize=14)
ax.text(5e9, 8e9, r"$x'$", color='red', fontsize=14)


ax.legend()


plt.show()
```

```python

velocities = [0.25, 0.5, 0.75]
colors = ['g', 'r', 'b']


fig, ax = plt.subplots(figsize=(7,7))


x_vals = np.linspace(0, 2e10, num=201, endpoint=True) - 1e10


for v, color in zip(velocities, colors):
    v_c = v * constants.c
    ct_vals_prime = ct_prime_axis(x_vals, v_c)
    ct_vals_x_prime = x_prime_axis(x_vals, v_c)

    ax.plot(x_vals, ct_vals_prime, linestyle='--', color=color, label=rf"$ct', v={v}c$")
    ax.plot(x_vals, ct_vals_x_prime, linestyle='--', color=color, label=rf"$x', v={v}c$")


ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


ax.set_aspect('equal')
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])


ax.set_xlabel(r"$x$ (m)", fontsize=14, color="black")
ax.set_ylabel(r"$ct$ (m)", fontsize=14, color="black")


ax.legend()
plt.show()
```

```python

velocities_gray = [0.5, 0.75, 0.875, 0.9375]
colors_gray = ['0.7', '0.5', '0.3', '0.1']


fig, ax = plt.subplots(figsize=(7,7))

x_vals = np.linspace(0, 2e10, num=201, endpoint=True) - 1e10


for v, color in zip(velocities_gray, colors_gray):
    v_c = v * constants.c
    ct_vals_prime = ct_prime_axis(x_vals, v_c)
    ct_vals_x_prime = x_prime_axis(x_vals, v_c)

    ax.plot(x_vals, ct_vals_prime, linestyle='--', color=color, label=rf"$ct', v={v}c$")
    ax.plot(x_vals, ct_vals_x_prime, linestyle='--', color=color, label=rf"$x', v={v}c$")

ax.plot(x_vals, x_vals, 'k-', label="Light Cone (c)")
ax.plot(x_vals, -x_vals, 'k-', label="Light Cone (-c)")


ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.set_aspect('equal')
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])


ax.set_xlabel(r"$x$ (m)", fontsize=14, color="black")
ax.set_ylabel(r"$ct$ (m)", fontsize=14, color="black")


ax.legend()
plt.show()
```
