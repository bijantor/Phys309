```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants


plt.rcParams.update({
    "font.size": 16,
    "axes.labelsize": 16,
    "font.family": "sans-serif",
    "font.sans-serif": "Arial"
})

q = -constants.e
epsilon_0 = constants.epsilon_0


r_prime = np.linspace(1e-10, 1e-8, 500)


E_prime = (1 / (4 * np.pi * epsilon_0)) * (q / r_prime**2)


plt.figure(figsize=(10, 7))
plt.plot(r_prime * 1e9, E_prime, label="Electric Field ($E'$)", color="red")
plt.title("Electric Field ($r'$)")
plt.xlabel("Distance $r'$ (nm)")
plt.ylabel("Electric Field $E'$ (N/C)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants


plt.rcParams.update({
    "font.size": 16,
    "axes.labelsize": 16,
    "font.family": "sans-serif",
    "font.sans-serif": "Arial"
})

c = constants.c


f = np.linspace(0, 1, 600)


c_prime = c * (1 - f)


plt.figure(figsize=(10, 6))
plt.plot(f, c_prime / 1e6, label=r"Wrong Speed of Light ($c'$)", color="red")
plt.title("Wrong Speed of Light ($c'$) vs Fraction of Velocity ($f = v/c$)")
plt.xlabel("Fraction of Velocity $f = v/c$")
plt.ylabel("Wrong Speed of Light $c'$ (x $10^6$ m/s)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants


plt.rcParams.update({
    "font.size": 16,
    "axes.labelsize": 16,
    "font.family": "sans-serif",
    "font.sans-serif": "Arial"
})


q = -constants.e
epsilon_0 = constants.epsilon_0
f = 0.5


r_prime = np.linspace(1e-10, 1e-8, 500)


E_prime_standard = (1 / (4 * np.pi * epsilon_0)) * (q / r_prime**2)


epsilon_0_modified = epsilon_0 * (1 + f)


E_prime_modified = (1 / (4 * np.pi * epsilon_0_modified)) * (q / r_prime**2)


plt.figure(figsize=(12, 6))
plt.plot(r_prime * 1e9, E_prime_standard, label=" $E'(r')$ (using $\\epsilon_0$)", color="blue")
plt.plot(r_prime * 1e9, E_prime_modified, label=f"$E'(r')$ (using $\\epsilon_0(1+f)$, $f={f}$)", color="red", linestyle="--")
plt.title("Change of $E'(r')$ with Standard and Modified $\\epsilon_0$")
plt.xlabel("Distance $r'$ (nm)")
plt.ylabel("Electric Field $E'(r')$ (N/C)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants


plt.rcParams.update({
    "font.size": 16,
    "axes.labelsize": 16,
    "font.family": "sans-serif",
    "font.sans-serif": "Arial"
})


mu_0 = constants.mu_0
I = 1
R = 1
f = np.linspace(0, 1, 500)


mu_0_modified = mu_0 * (1 + f)


B_modified = (mu_0_modified * I) / (2 * R)

plt.figure(figsize=(12, 6))
plt.plot(f, B_modified * 1e6, label=r"Modified $B(f) = \frac{\mu_0(f) I}{2R}$", color="purple")
plt.title("Magnetic Field $B(f)$ at the Center of the Loop as a Function of $f = v/c$")
plt.xlabel("Fractional Velocity $f = v/c$")
plt.ylabel("Magnetic Field $B(f)$ (x $10^{-6}$ T)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline((mu_0 * I) / (2 * R) * 1e6, color="black", linewidth=0.8, linestyle="--", label="Standard $B$")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants


plt.rcParams.update({
    "font.size": 16,
    "axes.labelsize": 16,
    "font.family": "sans-serif",
    "font.sans-serif": "Arial"
})


mu_0 = constants.mu_0
epsilon_0 = constants.epsilon_0
I = 1
R = 1


E_standard = 1
B_standard = (mu_0 * I) / (2 * R)


f = np.linspace(0, 1, 500)


E_prime = E_standard / (1 + f)
B_prime = B_standard * (1 + f)


plt.figure(figsize=(12, 6))
plt.plot(f, E_prime, label=r"$E'(f)$", color="blue")
plt.axhline(E_standard, color="blue", linestyle="--", label="Standard $E$")

plt.plot(f, B_prime * 1e6, label=r"$B'(f)$ (x $10^{-6}$ T)", color="red")
plt.axhline(B_standard * 1e6, color="red", linestyle="--", label="Standard $B$ (x $10^{-6}$ T)")

plt.title("Comparison of $E'(f)$ and $B'(f)$ with Standard $E$ and $B$")
plt.xlabel("Fraction of Velocity $f = v/c$")
plt.ylabel("Field Values")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
