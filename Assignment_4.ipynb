```python
from scipy.constants import Rydberg, h, c, e

def photon_energy(Z, n):


    inv_lambda = Rydberg * Z**2 * (1/n**2 - 1/(n + 1)**2)  # in 1/m


    energy_joules = h * c * inv_lambda


    energy_eV = energy_joules / e

    return energy_eV
```

```python
import numpy as np


n_values = np.arange(1, 11)


Z_values = np.arange(1, 4)

print("n values:", n_values)
print("Z values:", Z_values)
```

```python



Zv, nv = np.meshgrid(Z_values, n_values, indexing='ij')


energies = photon_energy(Zv, nv)


print("Photon energy array (in eV):")
print(energies)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable



fig, ax = plt.subplots(1, 1, figsize=(6, 6))


plot1 = ax.scatter(nv, Zv, c=energies, cmap='viridis', s=100, edgecolor='black')


ax.set_aspect('equal')


ax.set_xlabel("n (quantum number)", fontsize=12)
ax.set_ylabel("Z (atomic number)", fontsize=12)


divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="8%", pad=0.1)

cbar = plt.colorbar(plot1, cax=cax)
cbar.ax.tick_params(labelsize=12)
cbar.set_label("Photon Energy (eV)", fontsize=12)

plt.show()
```
