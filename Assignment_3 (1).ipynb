```python
import numpy as np
import matplotlib.pyplot as plt


c = 3e8
v_p = 0.99 * c
theta_deg = 15
theta_rad = np.radians(theta_deg)


v_sc = np.linspace(0, 0.98 * c, 300)


u_x_A = (v_p + v_sc) / (1 + (v_p * v_sc) / c**2)


v_px = v_p * np.cos(theta_rad)
v_py = v_p * np.sin(theta_rad)


gamma = 1 / np.sqrt(1 - (v_sc / c)**2)
u_y_B = v_py / (gamma * (1 + (v_sc * v_px) / c**2))

plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
plt.plot(v_sc / c, u_x_A / c, label="Proton Speed (x-direction)")
plt.xlabel("Spacecraft Speed (v/c)")
plt.ylabel("Proton Speed Measured by Frank (v/c)")
plt.title("Part A: Relativistic Velocity Addition")
plt.grid(True)
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(v_sc / c, u_y_B / c, label="y-component of Proton Speed", color="orange")
plt.xlabel("Spacecraft Speed (v/c)")
plt.ylabel("y-component Measured by Frank (v/c)")
plt.title("Part B: Transverse Velocity Transformation")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#did things in km
c = 3e5


t = np.linspace(0, 4, 500)
ct = c * t


x_flash_1 = 0
x_flash_2 = 300000


x_jwst = 150000


x_probe_start = 500000
v_probe = -0.1 * c


light_right = x_flash_1 + c * t
light_left = x_flash_2 - c * t


jwst_worldline = np.full_like(t, x_jwst)


probe_worldline = x_probe_start + v_probe * t

plt.figure(figsize=(10, 6))


plt.plot(x_flash_1, 0, 'ro', label='Flash 1 (x = 0 km)')
plt.plot(x_flash_2, 0, 'bo', label='Flash 2 (x = 300000 km)')

plt.plot(light_right, ct, 'r--', linewidth=1)
plt.plot(light_left, ct, 'b--', linewidth=1)


plt.plot(jwst_worldline, ct, 'g-', label='JWST World Line (x = 150000 km)')


plt.plot(probe_worldline, ct, 'm-', label='Probe World Line (starts at x = 500000 km, v = -0.1c)')


plt.xlabel("Position x (km)")
plt.ylabel("ct (km)")
plt.title("2D Spacetime Diagram")
plt.legend()
plt.grid(True)
plt.xlim(-50000, 550000)
plt.ylim(0, max(ct))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

star_types = {
    "O5": 40000,
    "B5": 15000,
    "A5": 8500,
    "F5": 6500,
    "G5": 5500,
    "K5": 4500,
    "M5": 3200
}


b = 2.898e-3


wavelengths_nm = {star: (b / T) * 1e9 for star, T in star_types.items()}

plt.figure(figsize=(10, 6))
plt.plot(list(star_types.keys()), list(wavelengths_nm.values()), marker='o')

for star, wl in wavelengths_nm.items():
    plt.text(star, wl + 10, f"{wl:.0f} nm", ha='center')

plt.title("Wavelength of Maximum Intensity for Different Star Types")
plt.xlabel("Star Type")
plt.ylabel("Peak Wavelength (nm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#decided to do in eV
h = 4.135667696e-15


phi_al = 4.08
phi_au = 5.10


frequencies = np.linspace(7e14, 1.5e15, 500)


ke_al = h * frequencies - phi_al
ke_au = h * frequencies - phi_au

ke_al = np.where(ke_al > 0, ke_al, 0)
ke_au = np.where(ke_au > 0, ke_au, 0)

plt.figure(figsize=(10, 6))
plt.plot(frequencies * 1e-14, ke_al, label="Aluminum", color="blue")
plt.plot(frequencies * 1e-14, ke_au, label="Gold", color="gold")

plt.xlabel("Frequency (10 Hz)")
plt.ylabel("Photoelectron Kinetic Energy (eV)")
plt.title("Photoelectric Effect: Kinetic Energy vs. Frequency")
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
